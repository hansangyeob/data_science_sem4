```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('USA_cars_datasets.csv')
```

```python
df
```

```python
df_subset = df[['color']]
```

```python
df_subset
```

```python
df_subset = pd.get_dummies(df_subset)
```

```python
df_subset
```

```python
df_subset['brand'] = df['brand']
```

```python
df_subset = df_subset.groupby('brand').sum()
```

```python
df_subset
```

```python
import matplotlib.pyplot as plt
width = 0.5       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots(figsize=(20,5))

ax.bar(df_subset.index, df_subset['color_black'], width, label='Black')
ax.bar(df_subset.index, df_subset['color_blue'], width,
       label='Blue')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()
plt.show()
```
