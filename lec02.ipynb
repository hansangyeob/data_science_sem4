```python
# retrieving financial data with investpy
import investpy

# get Apple stock prices
df = investpy.get_stock_historical_data(stock='AAPL', country='United States', from_date='01/01/2019', to_date='01/01/2020')
df.head(10)
```

```python
# get Google stock price
df_GOOGL = investpy.get_stock_historical_data(stock='GOOGL', country='United States', from_date='01/01/2019', to_date='01/01/2020')
df_GOOGL.head()
```

```python
# merge 2 datasets
merged_df = df.merge(df_GOOGL, left_on='Date', right_on='Date')
merged_df.head()
```

```python
# cleaning text string examples
code1 = "FR"
code2 = "   FR "
print(code1 == code2)
```

```python
# remove white spaces
code2 = code2.strip()
print(code1 == code2)
```

```python
# to lower case
code2 = code2.lower()
print(code2)
```

```python
# to upper case
code2 = code2.upper()
print(code2)
```

```python
# replace text
desc = "<h1>This description is for demo only.</h1> <p>The HTML tags should be remove.</p>"
desc = desc.replace("<h1>", "")
print(desc)
```

```python
# strip html beaiutifulsoup
import re
html_strip = re.compile(r'<.*?>')
desc = html_strip.sub('', desc)
print(desc)
```

```python
# fill na
df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                   [3, 4, np.nan, 1],
                   [np.nan, np.nan, np.nan, 5],
                   [np.nan, 3, 2, 4]],
                  columns=list('ABCD'))
df
```

```python
# Interpolate data
df.interpolate()
```

```python
# fillna with 0
df.fillna(0)
```

```python
# fillna with bfill or ffill
df.fillna(method='bfill')
```

```python
# calculate the mean of each column
df.mean(axis=0)
```

```python
# fillna with mean
df['B'] = df['B'].fillna(df['B'].mean())
df
```

```python
# PLEASE don't do this fillna for the whole data frame, you should fillna for each column!!! 
df.fillna(df.mean(axis=0))
```

```python
# drop duplicates example
df = pd.DataFrame([["Yum", 2, np.nan, 0],
                   ["Yum", 2, np.nan, 0],
                   ["Yum", 4, np.nan, 1],
                   ["Yum", np.nan, np.nan, 5],
                   ["Ken", 3, 2, 4]],
                  columns=list('ABCD'))
df
```

```python
# all columns
df.drop_duplicates()
```

```python
# sub set column and keep = 'first' or 'last'
df = df.drop_duplicates(subset="A", keep="first")
```

```python
# export the data without index column
df.to_csv('sample.csv', index=False)
```
